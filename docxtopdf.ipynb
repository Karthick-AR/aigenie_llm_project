```python
from docx2pdf import convert
import os

base = r"D:\Workspace\projects\LLM\Contracts"
# Convert all .docx files in the folder to PDF (outputs placed next to each .docx)
convert(base)

# Print created pdf paths
pdfs = [os.path.join(base, f) for f in os.listdir(base) if f.lower().endswith('.pdf')]
for p in pdfs:
    print(p)
```
